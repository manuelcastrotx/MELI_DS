```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import os
```

```python
# Traverse the folder to make an histogram of data types

base_path = f"{os.getcwd()}/Job/MELI"
raw_docs_path = "docs_raw"

# Simple view
os.listdir(f"{base_path}/{raw_docs_path}")
```

```python
def list_ftypes(path):

    lst = os.listdir(path)
    ftypes = []    

    for f in lst:
        if os.path.isfile(f"{path}/{f}"):
            dtype = f.split(".")[-1]
            ftypes.append(dtype)        

        elif os.path.isdir(f"{path}/{f}"):            

            ftypes = ftypes + list_ftypes(f"{path}/{f}")

    return ftypes
```

```python
docs_raw_path = f"{base_path}/{raw_docs_path}"

files_formats = list_ftypes(docs_raw_path)
print(f"Total files: {len(files_formats)}")

labels, values = np.unique(files_formats, return_counts=True)

fig = go.Figure(data=[go.Pie(labels=labels, values=values)])

fig.update_layout(title = f'Distribution by data type in folder: {docs_raw_path}',
                  width = 600)

fig.show()
```
